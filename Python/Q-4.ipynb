```python
import requests
import pandas as pd

def file_convert(link, csv_file):
    response = requests.get(link)
    data = response.json()

    # Converting data into a list of dictionaries
    txt = []
    for item in data:
        year = item.get('year', '')
        year = (year.split('-')[0]) if '-' in year else year
        meteorite = {
            'Name of Earth Meteorite': item.get('name', ''),
            'ID': item.get('id', ''),
            'Meteorite': str(item.get('nametype', '')),
            'Name Type': item.get('recclass', ''),
            'Rec Class': item.get('recclass', ''),
            'Mass': float(item.get('mass', '')) if item.get('mass', '').replace('.', '', 1).isdigit() else 0.0,
            'Year': year,
            'Rec Lat': float(item.get('reclat', '')) if item.get('reclat', '').replace('.', '', 1).isdigit() else 0.0,
            'Rec Long': float(item.get('reclong', '')) if item.get('reclong', '').replace('.', '', 1).isdigit() else 0.0,
            'Coordinates': item.get('coordinates', '')
        }
        txt.append(meteorite)

    # Converting data into a dataframe
    df = pd.DataFrame(txt)

    # Saving dataframe as an Excel file
    df.to_csv(csv_file, index=False)
    return "File has been downloaded"
```

```python
link = "https://data.nasa.gov/resource/y77d-th95.json"
file = "nasa_data.csv"
```

```python
file_convert(link, file)
```
