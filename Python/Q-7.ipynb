```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("nasa_data.csv")
```

```python
df
```

### 1. Get all Earth meteorites fell before year 2000

```python
#Best visualizaton --> Bar plot to know the year & number of meteroid
year_counts = df['Year'].value_counts().sort_index()

# Plot the bar plot
plt.bar(year_counts.index, year_counts.values)
```

```python
#For the name of meteroids before 2000

meteroids_names = df[df['Year'] <2000]['Name of Earth Meteorite'].tolist()
print(f"The name of meteroids before are:{len(meteroids_names)}")
for i in meteroids_names:
    print(i)
```

### 2. Get all the earth meteroids co-ordinates who fell before year 1970

```python
#For earth coordinates name

meteroids_names = df[df['Year'] <1970]['Rec Lat'].tolist()
print(f"The Latitue of coorinate before 1970:{len(meteroids_names)}")
for i in meteroids_names:
    print(i)
```

### 3. Get all those mass whose mass is more then 1000 kg

```python
meteroids_names = df[df['Mass'] > 1000]['Name of Earth Meteorite'].tolist()
print(f"The meteroid name whose mass is less then 1000:{len(meteroids_names)}")
for i in meteroids_names:
    print(i)
```
