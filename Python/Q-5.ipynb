### Question 5 -
Write a program to download the data from the given API link and then extract the following data with
proper formatting
Link - http://api.tvmaze.com/singlesearch/shows?q=westworld&embed=episodes

```python
import requests
import pandas as pd
from bs4 import BeautifulSoup#for htmml parser which there in file

def file_convert(link, csv_file):
    response = requests.get(link)
    data = response.json()

    # Converting data into a list of dictionaries
    txt = []
    for item in data["_embedded"]["episodes"]:
        soup = BeautifulSoup(item["summary"], "html.parser")
        summary = soup.get_text().strip()

        episode = {
            "id": int(item["id"]),
            "url": item["url"],
            "name": item["name"],
            "season": int(item["season"]),
            "number": int(item["number"]),
            "type": item["type"],
            "airdate": item["airdate"],
            "airtime": item["airtime"],
            "runtime": float(item["runtime"]),
            "average_rating": float(item["rating"]["average"]),
            "summary": summary,
            "medium_image_link": item["image"]["medium"],
            "original_image_link": item["image"]["original"]
        }
        txt.append(episode)

    # Converting data into a DataFrame
    df = pd.DataFrame(txt)

    # Saving DataFrame as a CSV file
    df.to_csv(csv_file, index=False)
    return "File has been downloaded"
```

```python

link = "http://api.tvmaze.com/singlesearch/shows?q=westworld&embed=episodes"
csv_file = "westworld_episodes.csv"
```

```python
file_convert(link, csv_file)
```
