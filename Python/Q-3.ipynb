### Q-3 Write a program, which would download the data from the provided link, and then read the data and convert that into properly structured data and return it in Excel format

Link : https://raw.githubusercontent.com/Biuni/PokemonGO-Pokedex/master/pokedex.json

```python
import json, requests
import pandas as pd
```

```python
def file_convert(link, excel_file):
    
    #Data downloading
    response = requests.get(link)
    data = response.json()
    
    #converting into txt format
    txt = []
    for i in data["pokemon"]:
        pokemon = {
            'id': i['id'],
            'num': i['num'],
            'name': i['name'],
            'img': i['img'],
            'type': ', '.join(i['type']),
            'height': i['height'],
            'weight': i['weight'],
            'candy': i.get('candy', ''),
            'candy_count': i.get('candy_count', ''),
            'egg': i.get('egg', ''),
            'spawn_chance': i.get('spawn_chance', ''),
            'avg_spawns': i.get('avg_spawns', ''),
            'spawn_time': i.get('spawn_time', ''),
            'weakness': ', '.join(i.get('weaknesses', [])),
            'next_evolution': ', '.join([evolution['name'] for evolution in i.get('next_evolution', [])]),
            'prev_evolution': ', '.join([evolution['name'] for evolution in i.get('prev_evolution', [])])
        }
        txt.append(pokemon)
        
    #Converting data into dataframe
    df = pd.DataFrame(txt)
    #Into excel
    df.to_excel(excel_file, index = False)
    return f"File has been downloaded"
    
```

```python
link = "https://raw.githubusercontent.com/Biuni/PokemonGO-Pokedex/master/pokedex.json"
file = "pokemon.xlsx"
```

```python
file_convert(link, file)
```
