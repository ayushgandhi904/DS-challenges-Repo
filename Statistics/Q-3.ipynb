Ans- a)

To find the Dispersion in both & Interpret the data: To find dispersion, it requires range & standard deviance. It can be done by pandas functions

```python
#Lets rename both column for easy convection
df["bp_before"] = df[" Blood Pressure Before (mmHg)"]
df["bp_after"] = df[" Blood Pressure After (mmHg)"]

df = df.drop([" Blood Pressure Before (mmHg)", " Blood Pressure After (mmHg)"], axis= 1)
```

```python
df
```

```python
#Range 
range_before = df['bp_before'].max() - df['bp_before'].min()
range_after = df['bp_after'].max() - df['bp_after'].min()

print("Range (bp_before):", range_before)
print("Range (bp_after):", range_after)
```

```python
#Deviation
std_before = df['bp_before'].std()
std_after = df['bp_after'].std()

print("Standard Deviation (bp_before):", std_before)
print("Standard Deviation (bp_after):", std_after)
```

<font color=green>

1. For BP before range is about 28 & standard deviation is about 6.6, so it can be considered as normal distibution.
2. Similarly BP after, range is about 23 & standard deviation is about 6.9, deviation is some what high, so spread is somewhat high compare to bp_before


```python
#Lets visualize

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
bp_before = df['bp_before']
bp_after = df['bp_after']

# Plotting the curves
plt.figure(figsize=(10, 6))
sns.kdeplot(bp_before, label='bp_before')
sns.kdeplot(bp_after, label='bp_after')
plt.xlabel('Blood Pressure')
plt.ylabel('Density')
plt.title('Distribution of Blood Pressure')
plt.legend()
plt.show()
```

#### Its distribuiton curve is similar, but it can be noted it doesn't follow gaussian distribution.

Ans: b) Mean & confidential interval

```python
mean_before = np.mean(bp_before)
mean_after = np.mean(bp_after)
```

```python
print(f"Mean before: {mean_before}")
print(f"Mean after {mean_after}")
```

```python
#For conidence interval
import scipy.stats as stats
```

```python

# Assuming your DataFrame is named 'df'
bp_before = df['bp_before']

# Calculate mean and standard deviation for bp_before
mean_before = np.mean(bp_before)
std_before = np.std(bp_before)
n_before = len(bp_before)

# Calculate 95% confidence interval
ci_before = stats.t.interval(0.95, n_before-1, loc=mean_before, scale=std_before/np.sqrt(n_before))

# Plotting the histogram with distribution curve and a line at the 95% confidence interval
plt.figure(figsize=(10, 6))
sns.histplot(bp_before, kde=True)
sns.kdeplot(bp_before, color='r')
plt.axvline(ci_before[0], color='g', linestyle='--', label='5% Confidence Interval')
plt.axvline(ci_before[1], color='g', linestyle='--')
plt.xlabel('Blood Pressure (bp_before)')
plt.ylabel('Frequency')
plt.title('Distribution of Blood Pressure (bp_before) with 5% Confidence Interval')
plt.legend()
plt.show()


```

```python
# Assuming your DataFrame is named 'df'
bp_after = df['bp_after']

# Calculate mean and standard deviation for bp_after
mean_after = np.mean(bp_after)
std_after = np.std(bp_after)
n_after = len(bp_after)

# Calculate 95% confidence interval
ci_after = stats.t.interval(0.95, n_after-1, loc=mean_after, scale=std_after/np.sqrt(n_after))

# Plotting the histogram with distribution curve and a line at the 95% confidence interval
plt.figure(figsize=(10, 6))
sns.histplot(bp_after, kde=True)
sns.kdeplot(bp_after, color='r')
plt.axvline(ci_after[0], color='g', linestyle='--', label='5% Confidence Interval')
plt.axvline(ci_after[1], color='g', linestyle='--')
plt.xlabel('Blood Pressure (bp_after)')
plt.ylabel('Frequency')
plt.title('Distribution of Blood Pressure (bp_after) with 5% Confidence Interval')
plt.legend()

plt.show()
```

Ans: c)  Mean absolute deviation and Standard deviation 

```python
mad_before = np.mean(np.abs(bp_before - np.mean(bp_before)))
mad_after = np.mean(np.abs(bp_after - np.mean(bp_after)))
```

```python

print("Mean Absolute Deviation (MAD) - bp_before:", mad_before)
print("Mean Absolute Deviation (MAD) - bp_after:", mad_after)
```

```python
sd_before = np.std(bp_before)
sd_after = np.std(bp_after)
print("Standard Deviation (SD) - bp_before:", sd_before)
print("Standard Deviation (SD) - bp_after:", sd_after)
```

#### Interpretation 
- A lower MAD indicates that the data points are, on average, closer to the mean.
The MAD values for both columns are relatively close, suggesting that the average deviation from the mean is similar in both cases.

- The SD for "bp_after" is slightly higher than the SD for "bp_before," suggesting a slightly larger dispersion or variability in blood pressure after a certain event or intervention.

Ans: d)  correlation coefficient and significance of it at 1% level

```python
correlation_coefficient, p_value = stats.pearsonr(bp_before, bp_after)

print("Correlation Coefficient:", correlation_coefficient)
print("p-value:", p_value)

if p_value < 0.01:
    print("The correlation is significant at the 1% level of significance.")
else:
    print("The correlation is not significant at the 1% level of significance.")
```
